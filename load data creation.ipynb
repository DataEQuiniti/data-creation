## Load modules

```python
import pandas as pd
import numpy as np

from faker import Faker
import random
```

## Settings

```python
fake = Faker('en_GB')
fake.seed(42)

random.seed(42)
```

## Create data

```python
business_data = {}
for i in range(0, 10):
    business_data[fake.company()] = {
        'postcode' : fake.postcode(),
        'sic_code' : random.randint(1, 10),
        'head_count' : round(np.random.normal(loc=200, scale=10)),
        'loan_amount' : round(random.randint(1000, 10000) / 500.0) * 500.0,
        'loan_rate_min' : round(random.uniform(0.1, 2), 2),
        'loan_rate_max' : round(random.uniform(4, 6), 2),
        'turnover' : np.random.normal(loc=random.uniform(-100000, 100000), scale=random.uniform(0.1, 2)),
        'years_active' : random.randint(0, 10)
    }
pd.DataFrame.from_dict(business_data, orient='index').reset_index()
```

```python
loaners_data = {}
for i in range(0, 10):
    loaners_data[fake.company()] = {
        'postcode' : fake.postcode(),
        'sic_code' : random.randint(1, 10),
        'min_head_count' : round(np.random.normal(loc=100, scale=10)),
        'max_head_count' : round(np.random.normal(loc=400, scale=10)),
        'loan_amount_min' : round(random.randint(1000, 10000) / 500.0) * 500.0,
        'loan_amount_max' : round(random.randint(100000, 100000) / 500.0) * 500.0,
        'loan_rate' : round(random.uniform(0.1, 5), 2),
        'turnover_max' : np.random.normal(loc=random.uniform(-100000, -50000), scale=random.uniform(0.1, 2)),
        'turnover_min' : np.random.normal(loc=random.uniform(100000, 500000), scale=random.uniform(0.1, 2)),
        'years_active_min' : random.randint(0, 6),
        'years_active_max' : round(np.random.normal(loc=15, scale=6))
    }
pd.DataFrame.from_dict(loaners_data, orient='index').reset_index()
```
